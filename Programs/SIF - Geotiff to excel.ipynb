```python
from osgeo import gdal
import numpy as np
import pandas as pd

# Path to your GeoTIFF file
geotiff_path = '"C:\Users\aniru\Machine Learning\Space Hackathon\Group-2\Topic-2\UttarPradesh\LULC\UP_LULC_2012.tif"'

# Open the GeoTIFF file
dataset = gdal.Open(geotiff_path)

# Read the raster data into a numpy array
raster_array = dataset.ReadAsArray()

# Mapping of pixel values to features and RGB color values
class_mapping = {
    1: ('Built-up', '255,0,0'),
    2: ('Kharif Crop', '255,209,0'),
    3: ('Rabi Crop', '255,158,0'),
    4: ('Zaid Crop', '158,81,43'),
    5: ('Double/Triple Crop', '158,207,31'),
    6: ('Current Fallow', '245,245,219'),
    7: ('Plantation', '0,204,0'),
    8: ('Evergreen Forest', '0,94,0'),
    9: ('Deciduous Forest', '107,120,31'),
    10: ('Degraded/Scrub Forest', '115,184,43'),
    11: ('Littoral Swamp', '5,130,94'),
    12: ('Grassland', '184,234,120'),
    13: ('Shifting Cultivation', '158,31,235'),
    14: ('Wasteland', '209,181,133'),
    15: ('Rann', '191,191,191'),
    16: ('Waterbodies max', '94,209,242'),
    17: ('Waterbodies min', '0,158,222'),
    18: ('Snow Cover', '255,191,196'),
}

# Get the dimensions of the raster
rows, cols = raster_array.shape

# Create a grid of coordinates
x_coords = np.arange(dataset.GetGeoTransform()[0], dataset.GetGeoTransform()[0] + dataset.GetGeoTransform()[1] * cols, dataset.GetGeoTransform()[1])
y_coords = np.arange(dataset.GetGeoTransform()[3], dataset.GetGeoTransform()[3] + dataset.GetGeoTransform()[5] * rows, dataset.GetGeoTransform()[5])

# Create a meshgrid from the coordinates
x_mesh, y_mesh = np.meshgrid(x_coords, y_coords)

# Flatten the arrays
flat_x = x_mesh.flatten()
flat_y = y_mesh.flatten()
flat_data = raster_array.flatten()

# Map pixel values to features and RGB color values
feature_values = [class_mapping[value][0] for value in flat_data]
rgb_values = [class_mapping[value][1] for value in flat_data]

# Create a pandas DataFrame
df = pd.DataFrame({'X': flat_x, 'Y': flat_y, 'Value': flat_data, 'Feature': feature_values, 'RGB': rgb_values})

# Save the DataFrame to CSV
df.to_csv('output.csv', index=False)

# Close the GeoTIFF dataset
dataset = None
```
